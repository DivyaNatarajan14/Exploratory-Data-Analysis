#UNIVARIATE ANALYSIS

```python
for column in df.select_dtypes(include='object').columns:
    # Bar chart
    plt.figure(figsize=(2,1))
    sb.countplot(x=df[column])
    plt.title(f'Bar Chart of {column}')
    plt.xticks(rotation=45,ha='right') #ha='right' means horizontal alignment of text in right side
    plt.show()
```

```python
for column in df.select_dtypes(include='number').columns:
# Histogram
  plt.figure(figsize=(2,1))
  sb.histplot(df[column], kde=True, color='skyblue')#kde=True adds kernel density estimate plot
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

BAR CHART FOR DESCRIPTIVE STATISTICS

```python
for column in df.select_dtypes(include='number').columns:
# Bar chart for descriptive statistics
  summary_stats = df[column].describe()
  plt.figure(figsize=(2,1))
  sb.barplot(x=summary_stats.index, y=summary_stats.values, palette='viridis')
  plt.title(f'Descriptive Statistics of {column}')
  plt.ylabel('Value')
  plt.xticks(rotation=45,ha='right')
  plt.show()
```

KERNEL DENSITY PLOT

```python
for column in df.select_dtypes(include='number').columns:
# Kernel Density Plot-- smooth estimate of the distribution
  plt.figure(figsize=(2,1))
  sb.kdeplot(df[column], fill=True, color='orange')
  plt.title(f'Kernel Density Plot of {column}')
  plt.xlabel(column)
  plt.xticks(rotation=45)
  plt.show()
```

# ***BIVARIATE ANALYSIS***

Bivariate analysis involving a categorical and a numerical feature can be visualized using various techniques.
Box Plot by Category:

Visualizes the distribution of a numerical variable for each category of a categorical variable.

sns.boxplot(x='CategoricalColumn', y='NumericColumn', data=df, palette='Set2')

plt.title('Box Plot by Category')

plt.show()

Violin Plot by Category:

Similar to a box plot but includes a kernel density estimation for better distribution

sns.violinplot(x='CategoricalColumn', y='NumericColumn', data=df, palette='Set2')

plt.title('Violin Plot by Category')

plt.show()

Bar Plot with Error Bars:

Displays the mean or median of the numerical variable for each category, along with error bars.

sns.barplot(x='CategoricalColumn', y='NumericColumn', data=df, ci='sd', palette='Set2')

plt.title('Bar Plot with Error Bars')

plt.show()

Point Plot:

Similar to a bar plot but represents the point estimate and confidence interval.

sns.pointplot(x='CategoricalColumn', y='NumericColumn', data=df, ci='sd', palette='Set2')

plt.title('Point Plot')

plt.show()

Swarm Plot:

Represents individual data points, providing a more detailed view of the distribution.

sns.swarmplot(x='CategoricalColumn', y='NumericColumn', data=df, palette='Set2')

plt.title('Swarm Plot')

plt.show()

Faceted Histograms:

Displays histograms for numerical variables within each category using facets.

sns.histplot(data=df, x='NumericColumn', hue='CategoricalColumn', multiple='stack', palette='Set2',

bins=20)

plt.title('Faceted Histograms')

plt.show()

These visualizations allow you to explore the relationship between a categorical and a numerical variable. Customize the column names, palette, and other parameters based on your specific dataset and analysis objectives.

Use seaborn's barplot to plot the mean 'Attrition' for each category in the selected numeric column

```python
columns_to_skip = ['Attrition', 'EmployeeID', 'DailyRate','MonthlyRate']

for column in df.select_dtypes(include='number').columns:
    # Skip plotting specific columns
    if column in columns_to_skip:
        continue


    # Create a bar plot for the binary attrition
    plt.figure(figsize=(8, 5))
    plt.xticks(rotation=0)

    # Use seaborn's countplot to plot the count of 'Attrtion' for each category in the selected numeric column
    sb.countplot(data=df, x=column, hue='Attrition', palette='Set2')

    plt.title(f'{column} vs Attrition (Yes/No)')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='Attrition', labels=['No', 'Yes'])
    plt.xticks(fontsize=5)
    plt.show()
```

```python
# Categorical variables of interest
cat_column1 = 'Education'
cat_column2 = 'Attrition'
# Stacked Bar Plot
cross_tab = pd.crosstab(df[cat_column1], df[cat_column2])
cross_tab.plot(kind='bar', stacked=True, colormap='viridis', figsize=(5,3))
plt.title(f'Stacked Bar Plot of {cat_column1} by {cat_column2}')
plt.xlabel(cat_column1)
plt.ylabel('Count')
plt.show()
```

```python
 # Calculate correlations between all variables
num_list2 = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = num_list2.corr()
# Use heatmap for visualization
plt.figure(figsize=(20, 8))
mask = np.triu(np.ones_like(corr_matrix))
sb.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5,mask=mask )
plt.xticks(rotation=45,fontsize=10)
plt.yticks(rotation=20,fontsize=7,ha='right')
plt.title("Correlation Matrix Between All Variables")
plt.show()
```
